```python
'''Simple normalization'''


import re 

def normalize_text(s):
    if pd.isna(s):
        return ""
    s = str(s).lower()
    s=re.sub(r"\s+", " ", s)
    return s.strip()

df = short

df["produkty_clean"] = df["Produkty"].apply(normalize_text)
```

```python
df[["Produkty", "produkty_clean"]].head(3)
```

```python
'''Spliter function'''

import re

def split_products(s):
    if not s:
        return []
    
    # 1) insert a separator AFTER "(xN)" if another letter starts right away
    #    e.g. "... (x1)rurka ..." -> "... (x1) | rurka ..."
    s = re.sub(r"(\(x\d+\))(?=\S)", r"\1 | ", s)

    # 2) now split on " + ", on "zestaw" starts, and on our inserted separator
    parts = re.split(r"\s\+\s|\s\|\s|(?=\bzestaw\b)", s)

    return [p.strip() for p in parts if p.strip()]
```

```python
product_matrix = pd.read_csv("../data/raw/product_matrix.csv")

product_matrix.head(3)
```

```python
'''Applying spliter function that separates different products within the same order to the product matrix'''

# Step 1: split product matrix rows into single-product rows
product_matrix["produkty_clean"] = product_matrix["Produkty"].apply(normalize_text)

matrix_long = product_matrix.copy()

matrix_long["product_clean_single"] = matrix_long["produkty_clean"].apply(split_products)
matrix_long = matrix_long.explode("product_clean_single").reset_index(drop=True)

# Quick sanity check
matrix_long[["Produkty", "product_clean_single"]].head(10)
```

```python
STOP_TOKENS = {"outlet"}  # can expand later if needed

def split_products_matrix(s):
    parts = split_products(s)  # reuse your existing splitter

    cleaned = []
    for p in parts:
        p = p.strip()

        # drop pure tokens like "outlet"
        if p in STOP_TOKENS:
            continue

        # if something still starts with "outlet ", strip it
        p = re.sub(r"^outlet\s+", "", p).strip()

        if p:
            cleaned.append(p)

    return cleaned


# re-run explode using the matrix-aware splitter

matrix_long["product_clean_single"] = matrix_long["produkty_clean"].apply(split_products_matrix)
matrix_long = matrix_long.explode("product_clean_single").reset_index(drop=True)

matrix_long[["Produkty", "product_clean_single"]].head(12)
```

```python
# Step 2: dedupe the matrix rows BEFORE splitting/exploding
df_unique = matrix_long.copy()

# keep only unique product descriptions (after cleaning)
df_unique = df_unique.drop_duplicates(subset=["produkty_clean"]).reset_index(drop=True)

print("Original rows:", len(df))
print("Unique rows:", len(df_unique))

# rebuild matrix_long cleanly from unique rows
matrix_long = df_unique.copy()
matrix_long["product_clean_single"] = matrix_long["produkty_clean"].apply(split_products_matrix)
matrix_long = matrix_long.explode("product_clean_single").reset_index(drop=True)

matrix_long[["Produkty", "product_clean_single"]].head(12)
```

```python
df["product_parts"] = df["produkty_clean"].apply(split_products)

df[["produkty_clean", "product_parts"]].head(5)
```

```python
parts = df[["product_parts"]].explode("product_parts").rename(columns={"product_parts": "part_raw"}).copy()
parts["part_raw"] = parts["part_raw"].fillna("")

```

```python
'''Extracting quantity'''

import re

def get_qty(s):
    m = re.search(r"\(x(\d+)\)", s)
    return int(m.group(1)) if m else 1

parts["qty"] = parts["part_raw"].apply(get_qty)
```

```python
def strip_qty(s):
    return re.sub(r"\s*\(x\d+\)\s*", " ", s).strip()

parts["part_clean"] = parts["part_raw"].apply(strip_qty)
```
